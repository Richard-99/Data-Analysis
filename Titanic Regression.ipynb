```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
male = pd.get_dummies(df.sex).iloc[:, 1]
```

```python
df = pd.get_dummies(df, columns = ['embarked'])
```

```python
df
```

```python
df.drop('embarked_Q', axis = 1, inplace = True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop('cabin', axis = 1, inplace = True)
```

```python
df
```

```python
pclass=pd.get_dummies(df['pclass'],drop_first=True)
pclass
```

```python
df = pd.concat([df, pclass, male], axis = 1)
df
```

```python
df = df.drop(['pclass','name', 'sex', 'age', 'sibsp', 'parch', 'ticket', 'fare', 'boat', 'body', 'home.dest'], axis = 1)
df
```

```python
X = df.drop('survived', axis = 1)
Y = df["survived"]
Y.fillna(1,inplace=True)
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test, Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train, Y_train)
```

```python
prediction=logmodel.predict(X_test)
prediction
```

```python
pd.DataFrame(prediction,Y_test)
```

```python
pre=list(Y_test)
aft=list(prediction)
```

```python
from sklearn.metrics import confusion_matrix

het=pd.DataFrame(confusion_matrix(Y_test,prediction))
het
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,prediction)*100
```
